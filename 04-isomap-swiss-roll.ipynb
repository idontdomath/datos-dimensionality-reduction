```python
import pylab as pl

from sklearn import manifold, datasets
from sklearn.manifold import Isomap

X, color = datasets.samples_generator.make_swiss_roll(n_samples=4500)
```

### Visualizaci√≥n de Swiss Roll en 3D

```python
fig = pl.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap=pl.cm.Spectral)
pl.show()
```

```python
X.shape
```

###¬†ISOMAP

```python
Se puede utilizar ISOMAP, especificando:
    - Cantidad de vecinos a utilizar
    - M√©todo para calcular las distancias de todos contra todos: Floyd-Warshall o 
```

```python
embedding = Isomap(n_components=2, n_neighbors=5, path_method='FW')
X_transformed = embedding.fit_transform(X)
X_transformed.shape
```

```python
fig = pl.figure(figsize=(10,10))
ax = fig.add_subplot(111)
ax.scatter(X_transformed[:, 0], X_transformed[:, 1], c=color, cmap=pl.cm.Spectral)
pl.show()
```
